```python
import pandas as pd
import numpy as np
```

```python
dates_dim = pd.read_csv("./datasets/raw/date_data/dimdates.csv")
dates_dim.head()
```

```python
dates_dim.shape
```

```python
dates_dim.dtypes
```

```python
# Relevant columns:
# Date ID , Year, Month Short Name, Month Long Name, Calendar Month, Number of days in month, 
# Calendar Quarter, Number of days in Quarter, isLeapYear 

dates_dim = dates_dim.rename(columns={"Id":"Date_ID"})
dates_dim = dates_dim.drop(columns=['CalendarMonthStartDate', 
                        'CalendarMonthEndDate', 
                        'CalendarQuarterStartDate', 
                        'CalendarQuarterEndDate',
                        'CalendarYearEndDate', 
                        'CalendarYearStartDate'])

display(dates_dim.head())
display(dates_dim.dtypes)
```

```python
# Adding leap year indicator [1: is lead year, 0: not leap year]

dates_dim['LeapYear'] = [1 if i == 366 else 0 for i in dates_dim['CalendarNumberOfDaysInYear']]
display(dates_dim.head())
```

```python
len(dates_dim) # 18263
```

```python
display(max(dates_dim['CalendarYear']))
display(min(dates_dim['CalendarYear']))
```

```python
dates_dim = dates_dim[dates_dim['CalendarYear'] >= 2001]
dates_dim = dates_dim[dates_dim['CalendarYear'] <= 2023]
display(len(dates_dim))
display(max(dates_dim['CalendarYear']))
display(min(dates_dim['CalendarYear']))
```

```python
dates_dim.dtypes
```

```python
dates_dim.drop_duplicates(subset = ['CalendarMonth', 'CalendarYear'], inplace=True)
display(len(dates_dim))
display(dates_dim.head())
```

```python
dates_dim['Date'] = pd.to_datetime(dates_dim['Date'])
dates_dim.dtypes
```

```python
dates_dim['Date_ID'] = range(0, len(dates_dim))
display(dates_dim.head())
```

```python
dates_dim = dates_dim.reset_index(drop=True)
display(dates_dim.head())
```

```python
dates_dim.isna().sum()
```

```python
dates_dim = dates_dim.rename(columns={
    'MonthLongName': 'Month_LongName',
    'MonthShortName': 'Month_ShortName',
    'CalendarMonth': 'Month',
    'CalendarNumberOfDaysInMonth': 'NumberOfDaysInMonth',
    'CalendarQuarter': 'Quarter',
    'CalendarNumberOfDaysInQuarter': 'NumberOfDaysInQuarter',
    'CalendarYear': 'Year',
    'CalendarNumberOfDaysInYear': 'NumberOfDaysInYear'
})
```

```python
dates_dim.to_csv("./datasets/dates_dim.csv",  index = False)
```
