Load Crop Prices

```python
cropPricesDF = pd.read_csv("./datasets/raw/crop_data/crops_prices.csv")
cropPricesDF = cropPricesDF[['REF_DATE', 'GEO', 'Farm products','VALUE']]
cropPricesDF['REF_DATE'] = pd.to_datetime(cropPricesDF['REF_DATE'])
cropPricesDF.rename(columns={'REF_DATE': 'Date', 'GEO': 'Province', 'VALUE':'Price','Farm products':'Product'}, inplace=True)
replace_dict = {'Wheat (except durum wheat) [1121111]': 'Wheat', 'Rye [1151152]': 'Rye', 'Oats [115113111]': 'Oats', 'Barley [1151141]': 'Barley', 'Canola (including rapeseed) [113111]': 'Canola', 'Durum wheat [112111211]': 'Durum', 'Flaxseed [115122111]': 'Flaxseed'}
cropPricesDF['Product'] = cropPricesDF['Product'].replace(replace_dict)
cropPricesDF
```

```python
dataframes = [quebecDF,ontarioDF,saskatchewanDF,britishcolumbiaDF,ontarioDF,manitobaDF,albertaDF]
mergeColumns = ['Date','Province','Product']
finalVolume_DF = pd.concat(dataframes, ignore_index=True)
print(len(finalVolume_DF))
cropPricesDF = pd.merge(cropPricesDF, finalVolume_DF, how='left', on=mergeColumns)
#Convert 0 values to NaN for processing
cropPricesDF['Volume'] = cropPricesDF['Volume'].replace(0, np.nan)
cropPricesDF['Price'] = cropPricesDF['Price'].replace(0, np.nan)



# Count NaN values in the 'volume' and 'price' columns
nan_count_volume = cropPricesDF['Volume'].isna().sum()
nan_count_crop_prices = cropPricesDF['Price'].isna().sum()
print(f"Number of NaN values in 'crop_prices': {nan_count_crop_prices}")
print(f"Number of NaN values in 'volume': {nan_count_volume}")


# Drop rows where both price and Volume is NaN
cropPricesDF = cropPricesDF.dropna(subset=['Price', 'Volume'], how='all')





```

```python
#Incorporate forward and backward fill for missing crop_prices and volume

price_ffill = cropPricesDF['Price'].ffill()


price_bfill = cropPricesDF['Price'].bfill()

# Calculate the mean of ffill and bfill series
mean_fill_series = (price_ffill + price_bfill) / 2
# Calculate global mean for remaining NaN prices
global_mean = cropPricesDF['Price'].mean()


cropPricesDF['Price'] = cropPricesDF['Price'].fillna(mean_fill_series)

cropPricesDF['Price'] = cropPricesDF['Price'].fillna(global_mean)

#Repeat process for Volume
volume_ffill = cropPricesDF['Volume'].ffill()

# Backward fill (bfill) to fill NaNs
volume_bfill = cropPricesDF['Volume'].bfill()

# Calculate the mean of ffill and bfill series
mean_fill_series = (volume_ffill + volume_bfill) / 2
global_mean = cropPricesDF['Volume'].mean()

# Assign the mean of ffill and bfill to NaN values in the original column
cropPricesDF['Volume'] = cropPricesDF['Volume'].fillna(mean_fill_series)

cropPricesDF['Volume'] = cropPricesDF['Volume'].fillna(global_mean)
#Add Date_Id
reference_date = pd.Timestamp('2001-01-01')
cropPricesDF['Date_Id'] = cropPricesDF['Date'].apply(lambda x: ((x.year - reference_date.year) * 12 + x.month - reference_date.month))

#Add Crop_Id
mapping_dict = {'Wheat': 0, 'Durum': 1, 'Oats': 2, 'Barley': 3,'Rye': 4,'Flaxseed': 5,'Canola': 6}
cropPricesDF['Crop_Id'] = cropPricesDF['Product'].map(mapping_dict)
cropPricesDF

# zero_count_price = (cropPricesDF['Price'] == 0).sum()
# zero_count_volume = (cropPricesDF['Volume'] == 0).sum()
# nan_count_volume = cropPricesDF['Volume'].isna().sum()
# nan_count_crop_prices = cropPricesDF['Price'].isna().sum()
# print(f"Number of NaN values in 'crop_prices': {nan_count_crop_prices}")
# print(f"Number of NaN values in 'volume': {nan_count_volume}")
# print(zero_count_price)
# print(zero_count_volume)
```
