```python
import pandas as pd
import math
import numpy as np
```

Binning

Region Dimension

```python
region_df = pd.read_csv('datasets/region_dim.csv')
region_df.head()
```

```python
region_df.loc[region_df['Name_EN'] == 'British Columbia','GeographicRegion'] = 'British Columbia'
region_df.head(20)
```

```python
binned_df = pd.DataFrame()

for col in ['Area','Land','LandOfArea_Percent','Freshwater','FreshwaterOfArea_Percent']:
        binned_df[col + '_Binned'] = pd.qcut(region_df[col], q=math.ceil(math.sqrt(len(region_df[col]))),duplicates='drop')
        binned_df[col + '_Binned_Encoded'] = pd.qcut(region_df[col], q=math.ceil(math.sqrt(len(region_df[col]))),duplicates='drop',labels=False)

binned_df.head(20)
```

```python
def encode_pop_bins(pop,bins):
    return bins.index(pop)
```

```python
bins = [0, 1000000, 5000000, 10000000, 15000000, 50000000]

region_df['Population_Binned_Encoded'] = region_df['Population_Lower'].apply(encode_pop_bins,args=(bins,))
region_df.head(20)
```

Constuct Training Data Set

```python
crop_df = pd.read_csv("datasets/crop_dimension_encoded.csv")
region_df = pd.read_csv("datasets/region_dim_binned_encoded.csv")
date_df = pd.read_csv("datasets\dates_dim.csv")
climate_df = pd.read_csv("datasets\climate_dimension_binned.csv") 
fact_table = pd.read_csv("datasets/cropFactTable.csv")
```

```python
# drop columns from Crop
crop_df.drop(['Unnamed: 0','Common Name','Species','Family'],axis=1,inplace=True)
crop_df.head()
```

```python
region_df.head()
```

```python
# drop columns from Region
cols = list(range(2,19))
region_df.drop(region_df.columns[cols],axis=1,inplace=True)
region_df.head()
```

```python
cols = [1,3,5,7,9]
region_df.drop(region_df.columns[cols],axis=1,inplace=True)
region_df.head()
```

```python
date_df.drop(['Date','Month_LongName','Month_ShortName','NumberOfDaysInQuarter','NumberOfDaysInYear'],axis=1,inplace=True)
date_df.head()
```

```python
climate_df.head()
```

```python
for col in climate_df.columns:
    if 'Climate_ID' not in col and 'Encoded' not in col:
        climate_df.drop(col,axis=1,inplace=True)
```

```python
climate_df.head(10)
```

```python
fact_table.drop(['Price','Volume','Total_Value','Total_Value_Binned'],axis=1,inplace=True) 
fact_table.head()
```

```python
# Create training data set by merging all 5 tables together 
pd.options.display.float_format = '{:.2f}'.format
combined_df = fact_table.merge(date_df,on='Date_ID').merge(crop_df,on='Crop_ID').merge(climate_df,on='Climate_ID').merge(region_df,on='Region_ID')
combined_df.head()
```

```python
combined_df.to_csv('datasets/merged_dataset.csv')
```

```python
combined_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = combined_df.drop(['Total_Value_Binned_Encoded'],axis=1)
y = combined_df['Total_Value_Binned_Encoded']
```

```python
y[y.isnull() == True]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)

X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=1)
```

```python
# Homogenize the index values,
X_train.index = y_train.index
# Assign the columns.
train_df = X_train
train_df['Total_Value'] = y_train
train_df.head()
```

```python
# Homogenize the index values,
X_train.index = y_train.index
# Assign the columns.
train_df = X_train
train_df['Total_Value'] = y_train

# Homogenize the index values,
X_val.index = y_val.index
# Assign the columns
valid_df = X_val
valid_df['Total_Value'] = y_val

# Homogenize the index values,
X_test.index = y_test.index
# Assign the columns.
test_df = X_test
test_df['Total_Value'] = y_test

print(len(train_df))
print(len(valid_df))
print(len(test_df))

train_df.to_csv('datasets/merged/training_dataset.csv') 
valid_df.to_csv('datasets/merged/validate_dataset.csv') 
test_df.to_csv('datasets/merged/test_dataset.csv') 
```
