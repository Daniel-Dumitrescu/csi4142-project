```python
import pandas as pd
import math
```

Binning

Region Dimension

```python
region_df = pd.read_csv('datasets/region_dim.csv')
region_df.head()
```

```python
binned_df = pd.DataFrame()

for col in ['Area','Land','LandOfArea_Percent','Freshwater','FreshwaterOfArea_Percent']:
        binned_df[col + '_Binned'] = pd.qcut(region_df[col], q=math.ceil(math.sqrt(len(region_df[col]))),duplicates='drop')
        binned_df[col + '_Binned_Encoded'] = pd.qcut(region_df[col], q=math.ceil(math.sqrt(len(region_df[col]))),duplicates='drop',labels=False)

binned_df.head(20)
```

```python
def encode_pop_bins(pop,bins):
    return bins.index(pop)
```

```python
bins = [0, 1000000, 5000000, 10000000, 15000000, 50000000]

region_df['Population_Binned_Encoded'] = region_df['Population_Lower'].apply(encode_pop_bins,args=(bins,))
region_df.head()
```

One-hot Encoding

Crop Dimension

```python
# Crop Dimension
crop_df = pd.read_csv("datasets\CropDimension.csv")
crop_df.head()
```

```python
name_one_hot = pd.get_dummies(crop_df['Common Name'])
name_one_hot.head()
```

```python
species_one_hot = pd.get_dummies(crop_df['Species'])
species_one_hot.head()
```

```python
family_one_hot = pd.get_dummies(crop_df['Family'])
family_one_hot.head(10)
```

```python
crop_df = pd.concat([crop_df,name_one_hot,species_one_hot,family_one_hot], axis=1)
crop_df.head()
```

```python
crop_df.to_csv("datasets/crop_dimension_encoded.csv")
```
